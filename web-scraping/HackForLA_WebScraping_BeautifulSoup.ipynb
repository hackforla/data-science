Use BeautifulSoup when:
- Website content is static (no JavaScript rendering)
- Data is present in page source
- You need a fast and lightweight scraper

### Why?

---

````python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://books.toscrape.com/catalogue/page-1.html")
html_content = response.content

# Use Beautiful Soup to parse the HTML
soup = BeautifulSoup(html_content, "html.parser")
````

````python
# The title tag of the page
print(soup.title)
# Output: <title>Hacker News</title>

# The title of the page as string
print(soup.title.string)
# Output: Hacker News

# All links in the page
nb_links = len(soup.find_all(class_="product_pod"))
print(f"There are {nb_links} links in this page")
# Output: There are 94 links on this page

# Text from the page
print(soup.get_text())
````

## Step 5: Locate Elements
```python
books = soup.find_all("article", class_="product_pod")
```

### Why?
- Each book is wrapped in an `article` tag
- `class_` helps filter relevant elements

---


````python

books = soup.find_all(class_="product_pod")

# Check if books were found
if books:
    # Loop through the selected elements
    for book in books:
        # Print each books text content to the console
        print(book.text)
````

## Step 6: Extract Data Using Loop
```python
data = []

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    data.append([title, price])
```

### Why?
- Loops allow extraction of multiple records
- Attributes like `title` are stored in HTML tags

---

````python
scraped_data = []
if books:
    for book in books:
        data = {
            "URL": book.find("a")["href"],
            "title": book.find("h3").find("a")["title"],
            "price": book.find(class_="price_color").get_text(strip=True),
        }
        scraped_data.append(data)

# Print the output list
print(scraped_data)
````
