```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from mpl_toolkits.mplot3d import art3d
import matplotlib
```

```python
fig = matplotlib.figure.Figure(figsize=(16,16))
ax = fig.add_subplot(111, projection='3d')

# Set angle
ax.azim=65
ax.elev=10.5

# Remove axes labels
ax.set_axis_off()
```

```python
def addrect(x, y, length, height, z, zdir, color, alpha):
  rect = matplotlib.patches.Rectangle((x, y), length, height)
  ax.add_patch(rect)
  art3d.patch_2d_to_3d(rect, z=z, zdir=zdir)
  rect.set_facecolor(color)
  rect.set_alpha(alpha)

def addline(startx, endx, starty, endy, z, zdir, color='grey', alpha=1, **kwargs):
  line = matplotlib.lines.Line2D([startx, endx], [starty, endy], color=color, alpha=alpha, zorder=.1, **kwargs)
  ax.add_line(line)
  art3d.line_2d_to_3d(line, zs=z, zdir=zdir)
  line.set_alpha(1)

```

```python
# Add background
addrect(0, 0, 1, 1, 0, 'x', 'lightgrey', .25)
addrect(0, 0, 1, 1, 0, 'y', 'lightgrey', .25)
addrect(0, 0, 1, 1, 0, 'z', 'lightgrey', .25)

# Add background grid
numberoflines = 6 
for direction in ['x', 'y', 'z']:
  for linenumber in range(numberoflines):
    lspot = linenumber / (numberoflines - 1)
    addline(lspot, lspot, 0.0001, 1, 0, direction, color='lightgrey')
    addline(0.0001, 1, lspot, lspot, 0, direction, color='lightgrey')


# Add spines
addline(0, 0, 0, 1, 0, 'x', color='darkgrey', linewidth = 3)
addline(0, 1, 0, 0, 0, 'x', color='darkgrey', linewidth = 3)
addline(0, 1, 0, 0, 0, 'y', color='darkgrey', linewidth = 3)

# # Add arrows
# arrowheadtobodyratio = .05
# arrowkwargs = {
#     "length": (1 + (1.25 *arrowheadtobodyratio)),
#     "arrow_length_ratio": arrowheadtobodyratio,
#     "color": "black"
# }
# ax.quiver3D(0,0,0,0,0,1, **arrowkwargs)
# ax.quiver3D(0,0,0,0,1,0, **arrowkwargs)
# ax.quiver3D(0,0,0,1,0,0, **arrowkwargs)

fig
```

```python
stepout=.125
squaresize = 1 - stepout
darkredhex = '#800016'

# center line
addline(1, 1, stepout, 1, 1, 'x', color=darkredhex)

#backleft
addline(stepout, stepout, stepout, 1, 1, 'x', color=darkredhex)
#topleft
addline(1, stepout, 1, 1, 1, 'x', color=darkredhex)
#bottomleft
addline(1, stepout, stepout, stepout, 1, 'x', color=darkredhex)

#backright
addline(stepout, stepout, stepout, 1, 1, 'y', color=darkredhex)
#topright
addline(1, stepout, 1, 1, 1, 'y', color=darkredhex)
#bottomright
addline(1, stepout, stepout, stepout, 1, 'y', color=darkredhex)

fig
```

```python
# Add grid lines for left pane
numpslines = 8

#left
for ps in range(numpslines):
  psloc = stepout + (squaresize * (ps / (numpslines-1)))
  #left vertical
  addline(psloc, psloc, stepout, 1, 1, 'x', color=darkredhex)
  #left horizontal
  addline(stepout, 1, psloc, psloc, 1, 'x', color=darkredhex)

fig
```

```python
# Add lines for right pane for aligning in photoshop
numpslines = 8

#left
for ps in range(numpslines):
  psloc = stepout + (squaresize * (ps / (numpslines-1)))
  #right vertical
  addline(psloc, psloc, stepout, 1, 1, 'y', color=darkredhex)
  #right horizontal
  addline(stepout, 1, psloc, psloc, 1, 'y', color=darkredhex)


fig
```
