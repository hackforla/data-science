```python
df_crimes = df_crimes.iloc[:, 3: ]
```

```python
df_crimes.head()
```

```python
df_crimes.index = pd.to_datetime(df_crimes.index)
```

```python
print(df_crimes.shape)
print(df_crimes.head)
```

```python
crimes_count = pd.DataFrame(df_crimes.groupby('Crm Cd Desc').size().sort_values(ascending=False).rename('counts').reset_index())
```

```python
crimes_count.head()
```

```python
crimes_count.shape
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))


# Plot the total crimes
sns.set_color_codes("pastel")
sns.barplot(x="counts", y="Crm Cd Desc", data=crimes_count.iloc[:10, :],
            label="Total", color="b")

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="Type",
       xlabel="Crimes")
sns.despine(left=True, bottom=True)

# Add a legend and informative axis label
plt.show()
```
