# Web Scraping with Selenium (Dynamic Websites)
## Overview

This tutorial demonstrates how to scrape dynamic websites using Selenium in Python.

**Unlike BeautifulSoup, Selenium can:**

- Interact with JavaScript-rendered content

- Simulate real browser behavior

- Click buttons, scroll pages, and wait for elements to load

**By the end of this tutorial, you will be able to:**

- Launch a browser using Selenium

- Navigate to a webpage

- Locate elements using XPath

- Extract structured data

- Store scraped results in a DataFrame

# When to Use Selenium

**Use Selenium when:**

- Data is loaded dynamically using JavaScript

- Content does not appear in page source

- Pages require scrolling, clicking, or waiting

- APIs are unavailable

⚠️ Selenium is slower than BeautifulSoup but much more powerful.

## Step 2: Import Libraries

**Explanation**

- **webdriver:** Controls the browser

- **By:** Helps locate elements using XPath, ID, class, etc.

- **time:** Adds delays to allow content to load

- **pandas:** Converts data into a DataFrame

```python
#!pip install selenium webdriver-manager 

import pandas as pd

import time
from selenium import webdriver  #used to automate web browser interaction
from selenium.webdriver.chrome.service import service #This helps in managing the ChromeDriver service that Selenium uses to interact with the Chrome browser.
from webdriver_manager.chrome import ChromeDriverManager #This class automatically downloads and sets up the latest version of ChromeDriver

from selenium.webdriver.common.by import By  #The By class is used to locate elements on a web page
```

## Step 3: Launch the Browser

**What this does**

- Automatically downloads the correct ChromeDriver
- Launches a Chrome browser instance
- Eliminates manual driver setup

```python
#Get Chrome driver
driver = webdriver.Chrome()
```

## Step 4: Open the Target Website

**Why time.sleep()?**

- Allows JavaScript and page elements to load
- Prevents errors when elements are not immediately available

```python
# driver wait 10 seconds until the page loaded

driver.implicitly_wait(10)

# Enter to the site
driver.get('https://books.toscrape.com/catalogue/page-1.html');
time.sleep(5)
```

## Step 5: Create a Data Container

**Why a list?**

- Each book is one record
- Lists allow easy iteration and storage
- Ideal for conversion to DataFrame later

---

## Step 6: Locate All Book Elements

**Explanation**

- Each book is inside an **<article>** tag with class **product_pod**
- XPath allows precise selection of repeating elements
- **find_elements()** returns a list of matching elements

---
## What each field does

**get_attribute("title")**
- Extracts full book title
- More reliable than visible text

**get_attribute("href")**
- Extracts absolute link to book detail page

**.text**
- Extracts visible price text
- Already cleaned and readable


```python
##initialize dictionary with lists
book_info = {'Title':[],
            'BookURL':[],
            'Price':[],
            'CoverimageURL':[],
             'Rating':[],             
            }
# Iterate over each product div
for row in driver.find_elements(By.XPATH,"//article[@class ='product_pod']"):
    
    title = row.find_element(By.XPATH,"h3/a").get_attribute('title')
    bookurl = row.find_element(By.XPATH,"h3/a").get_attribute('href')
    price = row.find_element(By.XPATH,"div[@class='product_price']/p").text
    coverimageurl = row.find_element(By.XPATH,"div[@class='image_container']/a/img").get_attribute('src')
    rating = row.find_element(By.XPATH,"p").get_attribute('class')

 #print(title,bookurl,price,coverimageurl,rating)
    
    # Extract and append data to the dictionary
    
    book_info['Title'].append(title)
    book_info['BookURL'].append(bookurl)
    book_info['Price'].append(price)
    book_info['CoverimageURL'].append(coverimageurl)
    book_info['Rating'].append(rating)
    
# Convert the dictionary to a pandas DataFrame

bookDF = pd.DataFrame(book_info)

# Display the DataFrame

bookDF.head()    
```

## Common Selenium Issues and Fixes


| Issue                       | Solution                   |
| --------------------------- | -------------------------- |
| Element not found           | Increase wait time         |
| Page loads slowly           | Use `WebDriverWait`        |
| XPath breaks                | Re-inspect HTML            |
| Browser closes unexpectedly | Check driver compatibility |


##Best Practices for Selenium Scraping

- Always add delays
- Avoid aggressive scraping
- Use relative XPath (.//)
- Handle missing elements with try/except
- Respect robots.txt and Terms of Service
